```python
def find_row_type(row):
    if row['address'] == row['business_vicinity1']:
        return row['business_type1']
    if row['address'] == row['business_vicinity2']:
        return row['business_type2']
    if row['address'] == row['business_vicinity3']:
        return row['business_type3']
    if row['address'] == row['business_vicinity4']:
        return row['business_type4']
    if row['address'] == row['business_vicinity5']:
        return row['business_type5']
```

```python
def find_row_name(row):
    if row['address'] == row['business_vicinity1']:
        return row['business_name1']
    if row['address'] == row['business_vicinity2']:
        return row['business_name2']
    if row['address'] == row['business_vicinity3']:
        return row['business_name3']
    if row['address'] == row['business_vicinity4']:
        return row['business_name4']
    if row['address'] == row['business_vicinity5']:
        return row['business_name5']
```

```python
def find_fuzzy_row_type(row):
    i=95
    if fuzz.ratio(row['address'],row['business_vicinity1']) >= i:
        return row['business_type1']
    if fuzz.ratio(row['address'],row['business_vicinity2']) >= i:
        return row['business_type2']
    if fuzz.ratio(row['address'],row['business_vicinity3']) >= i:
        return row['business_type3']
    if fuzz.ratio(row['address'],row['business_vicinity4']) >= i:
        return row['business_type4']
    if fuzz.ratio(row['address'],row['business_vicinity5']) >= i:
        return row['business_type5']
```

```python
def find_fuzzy_row_name(row):
    i=85
    if fuzz.ratio(row['address'],row['business_vicinity1']) >= i:
        return row['business_name1']
    if fuzz.ratio(row['address'],row['business_vicinity2']) >= i:
        return row['business_name2']
    if fuzz.ratio(row['address'],row['business_vicinity3']) >= i:
        return row['business_name3']
    if fuzz.ratio(row['address'],row['business_vicinity4']) >= i:
        return row['business_name4']
    if fuzz.ratio(row['address'],row['business_vicinity5']) >= i:
        return row['business_name5']
```
